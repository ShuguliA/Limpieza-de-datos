## Error 4 y 5: Redundancia de datos y  Columnas redundantes.

**Solución**

Para esta actividad la mejor solución sería eliminar las columas 3 y 4,puesto que la información ya estaría en la columna 5 y estaría duplicada,provocando una redundancia de datos.
